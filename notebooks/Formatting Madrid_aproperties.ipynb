```python
columns = [
    'id', 'sq_mt_built', 'n_rooms', 'n_bathrooms', 'is_exact_address_hidden',
    'floor', 'is_floor_under', 'rent_price', 'is_rent_price_known',
    'buy_price', 'buy_price_by_area', 'is_buy_price_known', 'is_renewal_needed',
    'is_new_development', 'built_year', 'has_central_heating', 'has_individual_heating',
    'has_ac', 'has_fitted_wardrobes', 'has_lift', 'is_exterior', 'has_terrace',
    'has_storage_room', 'energy_certificate', 'has_parking', 'is_parking_included_in_price',
    'parking_price', 'is_orientation_north', 'is_orientation_west', 'is_orientation_south',
    'is_orientation_east', 'neighborhood', 'District', 'HouseType'
]
```

```python
# List of keywords to check
keywords_housetype = ['flat', 'house', 'building', 'penthouse', 'apartment', 'villa', 'lofts', 'loft', 'land', 'duplex', 'new development', 'homes', 'chalet', 'Piso', 'Property', 'Plot']

# Function to extract house type from heading
def extract_house_type(heading):
    heading_lower = heading.lower()
    for keyword in keywords_housetype:
        if keyword.lower() in heading_lower:
            return keyword.capitalize()
    return 'Unknown'

# Apply the function to the heading column and create the HouseType column
data['HouseType'] = data['heading'].apply(extract_house_type)

# Display the DataFrame
data.head()
```

```python
# Display the rows where HouseType is 'Unknown'
unknown_house_type_rows = data[data['HouseType'] == 'Unknown']
print("Rows with unknown house types: ")
print(len(unknown_house_type_rows))
unknown_house_type_rows
```

```python
data.loc[data['HouseType'] == 'Unknown', 'HouseType'] = 'Apartment'
```

```python
# Function to clean area column
def clean_area(area_str):
    try:
        # Check if area_str is already float (possibly due to previous conversion)
        if isinstance(area_str, float):
            return None  # Return None for float types

        # Process range case
        if '-' in area_str:
            range_vals = area_str.split('-')
            avg_val = (float(range_vals[0].replace(' m2', '')) + float(range_vals[1].replace(' m2', ''))) / 2
            return avg_val
        else:
            # Otherwise, convert to float after removing ' m2'
            return float(area_str.replace(' m2', ''))
    except (ValueError, AttributeError):
        return None  # Handle any conversion errors gracefully

# Apply cleaning function to area column
data['sq_mt_built'] = data['area'].apply(clean_area)

# Drop the original 'area' column
data.drop(columns=['area'], inplace=True)

# Display the updated DataFrame
data
```

```python
# Remove Euro sign from Price column
# Remove Euro sign and 'from' from Price column
data['buy_price'] = data['Price'].str.replace('€', '').str.replace('from', '').str.strip()
data.drop(columns=['Price'], inplace=True)
# Display the updated DataFrame
data
```

```python
# Remove dash and take text after it using split and strip
data['location'] = data['location'].apply(lambda x: x.split('-')[-1].strip() if '-' in x else x.strip())
```

```python
data
```

```python
# Function to extract built year
def extract_built_year(content):
    if 'built in' in content.lower():
        # Find index of 'built in' and extract year
        idx = content.lower().find('built in') + len('built in')
        year_str = content[idx:idx+4]  # Assuming year format is YYYY
        try:
            year = int(year_str)
            return year
        except ValueError:
            return None
    else:
        return None

# Apply function to create built_year column
data['built_year'] = data['content'].apply(extract_built_year)

# Display the updated DataFrame
data
```

```python
def extract_features(content):
    has_storage_room = 1 if 'storage room' in content.lower() else 0
    has_parking = 1 if 'parking' in content.lower() else 0
    has_terrace = 1 if 'terrace' in content.lower() else 0
    has_fitted_wardrobes = 1 if 'wardrobes' in content.lower() else 0
    return pd.Series([has_storage_room, has_parking, has_terrace, has_fitted_wardrobes], index=['has_storage_room', 'has_parking', 'has_terrace', 'has_fitted_wardrobes'])

# Apply function to create/update columns
data[['has_storage_room', 'has_parking', 'has_terrace', 'has_fitted_wardrobes']] = data['content'].apply(extract_features)

# Display the updated DataFrame
data
```
